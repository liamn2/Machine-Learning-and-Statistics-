<span style="font-family:Times New Roman; font-size:1.5em;">ANOVA can use many factorials in different versions. For this project, I will be looking at one-way analysis of variance, that is, analysis of variance of a single factorial with measurement repetition. We will use this as an extension of the t-test for independent samples. An example for which we could use this test would be in average salary between staff departments. As the avarage salary is independent, this makes them ideal for use in analysis of variance. Our dependent variable is staff seniority, that is that the test subject's salary is dependent on their seniority in their workplace.

<span style="font-family:Times New Roman; font-size:1.5em;">One-Way Analysis of Variance (ANOVA) tells you if there are any statistical differences between the means of three or more independent groups. ANOVA stands for Analysis of Variance. Itâ€™s a statistical test that was developed by Ronald Fisher in 1918 and has been in use ever since. Put simply, ANOVA tells you if there are any statistical differences between the means of three or more independent groups. For our test, I have taken an example from GeeksForGeeks, where we test 4 diferrent oil types on 5 of the same car each for 100km straight maximum (No car met this value). 

```python
from scipy.stats import f_oneway

performance1 = [89, 89, 88, 78, 79]  # Data as arrays
performance2 = [93, 92, 94, 89, 88]
performance3 = [89, 88, 89, 93, 90]
performance4 = [81, 78, 81, 92, 82]

f_oneway(performance1, performance2, performance3, performance4)  # Array fed directly into scipy.stats f_oneway function.
```

<span style="font-family:Times New Roman; font-size:1.5em;">The F statistic and p-value turn out to be equal to 4.625 and 0.016336498 respectively. Since the p-value is less than 0.05 hence we would reject the null hypothesis. This implies that we have sufficient proof to say that there exists a difference in the performance among four different engine oils. We will next use the same data set in the statsmodels Python module. 

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

performance1 = [89, 89, 88, 78, 79]                   # Data as arrays
performance2 = [93, 92, 94, 89, 88]
performance3 = [89, 88, 89, 93, 90]
performance4 = [81, 78, 81, 92, 82]

# Converting data into Pandas dataframe.
df = pd.DataFrame({'val':np.concatenate([performance1,performance2,performance3,performance4]),
             'data':np.repeat(['A','B','C','D'],[len(performance1),len(performance2),len(performance3),len(performance4)])})

df.head()

mod = ols('val ~ data',data=df).fit()  # Create model by fitting dataframe.

g = sm.stats.anova_lm(mod, typ=1) # Perform ANOVA test

print(g)

plt = g.plot()
type(plt)
```

<span style="font-family:Times New Roman; font-size:1.5em;">Using the scipy.stats module is a lot easier to use as we do not have to convert our array data into a pandas dataframe. Instead, we can directly feed arrays into scipy.stats f_oneway function to get our ANOVA result. That being said, statsmodels is for a more advanced audience in data analytics and statistics as the results are verbose and exhaustive. 

```python
import matplotlib
import matplotlib.pyplot as plt

x_coordinates = [1, 2, 3, 4, 5]

y1_coordinates = [89, 89, 88, 78, 79]
y2_coordinates = [93, 92, 94, 89, 88]
y3_coordinates = [89, 88, 89, 93, 90]
y4_coordinates = [81, 78, 81, 92, 82]

plt.plot(x_coordinates, y1_coordinates) # plot first line
plt.plot(x_coordinates, y2_coordinates) # plot second line
plt.plot(x_coordinates, y3_coordinates) # plot third line
plt.plot(x_coordinates, y4_coordinates) # plot fourth line

plt.xlabel(f"Car (1, 2, 3, 4, 5) respectively")
plt.ylabel("Kilomtres (Km) driven")
plt.title(f"Kilometres driven by 5 cars using 4 different oil types")
  
plt.show()
```
